## Основные теоретические положения

### Действие группы на множество

Пусть $G$ - группа, $X$ - множество. Говорят, что $G$ действует на $X$, если задана операция $G \times X \to X$, обладающая следующими свойствами:  
$\forall x \in X$ и $\forall g,h \in G$  
1) $g(hx) = (gh)x$  
2) $ex = x$


### Орбиты и неподвижные точки

Орбитой $x \in X$ под действием $G$ называется множество из элементов, в которые переходят $x$ под действием $G$.  
$Gx = \{gx | g \in G\}$.  
Длина орбиты - количество элементов в орбите. Любые две орбиты либо не пересекаются, либо совпадают.

Множество неподвижных точек элемента $g \in G$ - это такие $x \in X$, которые под действием элемента не изменяются.  
$X^g = \{x\in X | gx = gx\}$

### Лемма Бёрнсайда

Количество орбит действия группы $G$ на множестве $X$:  
$N = \frac{1}{|G|}\sum_{g \in G} |X^{g}|$

### Раскраски

Пусть $I$ - произвольное множество, $C$ - множество цветов. Раскраской $I$ называется отображение $I$ в $C$. Множество всех раскрасок обозначим как $C^I$.  

Количество раскрасок из $C^{I}$, не меняющиеся под действием $g \in G$ равно $|C|^{c(\varphi(g))}$, где:  
$c(\varphi(g))$ - количество независимых циклов в перестановке $\varphi(g)$ (вклюачя циклы длины 1).

Симметрии квадрата описывает группа $D_8$ (иногда записывают как $D_4$), порядок которой - 8. Перечислим все симметрии:  
1) $e$ - тождественное отображение,  
2) $r$ - поворот на 90 градусов,  
3) $r^2$ - поворот на 180 градусов,  
4) $r^3$ - поворот на 270 градусов,  
5) $s$ - отражение относительно вертикальной оси,  
6) $sr$ - отражение относительно побочной диагонали,   
7) $sr^2$ - отражение относительно горизонтальной оси,  
8) $sr^3$ отражение относительно главной диагонали.  

Необходимо для каждого элемента из $D_8$ найти количество независимых циклов.  
Количество клеточек в квадрате - 49. Заметим, что центральная клеточка никогда не будет покидать своего места, и будет являться циклом длины 1 во всех симметриях.  
1) $e$: очевидно, что количество независимых циклов равно количеству клеточек - $49$.  
2) $r$: легко заметить, что все циклы, кроме одного, будут состоять из 4 элементов, т.к. поворот на 90 градусов "смещает" четвёрки клеточек на 1 позицию; последний цикл, длины 1, состоит из центрального элемента. Итого $1+(49-1)/4=13$ циклов.  
3) $r^2$:  если проследить за движением клеточек, можно выделить пары, переходящие друг в друга относительно центрального элемента. Сам центральный остаётся на месте. Итого $1+(49-1)/2=25$ циклов.  
4) $r^3$: здесь применимы те же рассуждения, что и для $r$. $13$ циклов.  
5) $s$: нетрудно увидеть, что большинство элементов можно выделить в пары, которые переходят друг в друга при вертикальном отражении. Остальные элементы, находящиеся прямо на вертикальной оси, остаются на местах, образуют 7 циклов длины 1. Итого $7+(49-7)/2=28$ циклов.  
6) 7) 8) $sr, sr^2, sr^3$: здесь применимы те же рассуждения, что и для $s$ - меняется лишь ось симметрии. Итого 3 раза по $28$ циклов.  


Таким образом, согласно лемме Бёрнсайда,  
$N = \frac{1}{8}(5^{49} + 2*5^{13} + 5^{25} + 4*5^{28})$ - количество различных способов составить мозайку размера 7x7 из стёкол 5 различных цветов с точностью до поворотов и симметрий.
