## Основные теоретические положения

#### Определение действия группы на множестве

Пусть $\mathbb{G}$ - группа, $e$ - нейтральный элемент группы $\mathbb{G}$, $\mathbb{X}$ - множество.

$\mathbb{G}$ действует на $\mathbb{X}$, если задана операция $\mathbb{G} \times \mathbb{X} \rightarrow \mathbb{G}$ такая, что $\forall x \in \mathbb{X}$, $g, h \in \mathbb{G}$ выполняется:
1. $e x = x$
2. $g (h x) = (g h) x$

#### Определение орбиты

Орбитой $x \in \mathbb{X}$ под действием группы $\mathbb{G}$ называется множество $\mathbb{G} x=\{g x | g \in \mathbb{G}\} = Orb(x)$. 

Длина орбиты равна $|\mathbb{G} x|$.

#### Определение неподвижной точки

Неподвижными точками элемента $g \in \mathbb{G}$ называются такие $x \in \mathbb{X}$, для которых выполняется: $g x = x$. 

Множество неподвижных точек элемента $g \in \mathbb{G}$ обозначается как $\mathbb{X}^g$.

#### Лемма Бернсайда

Количество орбит действия группы $\mathbb{G}$ на $\mathbb{X}$ равно $\frac{1}{|\mathbb{G}|} \sum_{g \in \mathbb{G}}{|\mathbb{X}^g|}$. 

#### Определение раскраски

Пусть $I$ - произвольное множество, $\mathbb{C}$ - множество цветов. 

Раскраской множества $I$ называется функция $\varphi: I \rightarrow \mathbb{C}$. 

$\mathbb{C}^I$ - множество всех раскрасок. 

$|\mathbb{C}^I|=|\mathbb{C}|^{|I|}$ - количество всех раскрасок.

#### Определение 5

Пусть $\mathbb{G}$ - группа, действующая на множестве $I$. 

Тогда формула $gf(i)=f(g^{-1}i)$, где $g \in \mathbb{G}$, $i \in  I$, $f \in \mathbb{C}^{I}$, задает действие $\mathbb{G}$ на множество $\mathbb{C}^{I}$. 

#### Теорема

Для элемента $g \in \mathbb{G}$ и функции $f \in \mathbb{C}^{I}$ эквивалентны следующие условия:
1. $gf=f$
2. $f(i) = f(g^n i)$, $\forall i \in I$, $\forall n \in \mathbb{Z}$
3. Если $i$, $j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i)=f(j)$.

#### Следствие 1

Количество раскрасок из $\mathbb{C}^{I}$, которые сохраняют данный элемент $g \in \mathbb{G}$, равно $|\mathbb{C}|^{c(\varphi(g))}$, где $c(\varphi)$ - количество независимых циклов в записи перестановки $\varphi$.

#### Лемма 1

Пусть $\tau$ является циклом длины $k$, а $НОД(k, n) = d$.

Тогда перестановка $\tau^n$ является произведением d циклов длины $\frac{k}{d}$.

### Аналитическое решение задачи о раскрасках

Дано множество из 6 цветов $\mathbb{C}$ и множество секторов круга $I = \{x_1, x_2, \ldots, x_{70}\}$, $x_i \in \mathbb{C}$. Необходимо найти количество различных способов раскрасить сектора круга с точностью до поворота круга.

Всего вариантов раскраски секторов круга: $6^{70}$. На множество $\mathbb{C}^I$ всех раскрасок действует группа поворотов $\mathbb{G}$ с операцией композиции. Раскраски, совмещаемые друг с другом каким-нибудь преобразованием из группы $\mathbb{G}$, будут задавать одинаковый круг. При этом раскраски будут задавать одинаковый круг тогда и только тогда, когда они лежат в одной орбите под действием группы $\mathbb{G}$. Таким образом задача сводится к нахождению количества орбит действия группы $\mathbb{G}$ на множество $\mathbb{C}^I$.

Тогда $\mathbb{G} = \{e=\tau^{0}, \tau^{1}, \tau^{2}, \ldots, \tau^{68}, \tau^{69}\}.$

Найдем $НОД(70, k)$ для всех $k \in \{0, 1, 2, 3, \ldots, 69\}$:

```python
gcds = dict()

for k in range(70):
    gcd = GCD(70, k)
    
    if gcd in gcds.keys():
        gcds[gcd].append(k)
    else:
        gcds[gcd] = [k]
        
for pair in gcds.items():
    print("НОД(70, k) = {} для k, принадлежащих {}. Всего элементов: {}.".format(pair[0], pair[1], len(pair[1])))
    
```

### Реализация набора классов и функций для графического отображения раскраски по заданному номеру 

Реализуем набор классов для графического отображения раскраски по заданному номеру:

```python
from random import randint
from enum import Enum

# Перечисление, содержащее цвет (множество цветов)
class Color(Enum):
    RED = (0.95, 0, 0)
    GREEN = (0, 0.7, 0)
    BLUE = (0.2, 0.2, 1)
    YELLOW = (1, 1, 0)
    ORANGE = (1, 0.5, 0)
    PURPLE = (0.6, 0, 0.6)

# Класс сегмента круга  
class Segment:
    def __init__(self, number, color):
        self.number = number
        self.color = color
        
# Класс круга    
class Circle:
    def __init__(self, segments_number, number = None):
        self.segments_number = segments_number
        color_list = list(Color)
        
        if number is None:
            self.segments = [Segment(s + 1, color_list[randint(0, 5)]) for s in range(segments_number)]
        else:
            self.segments = [Segment(s + 1, color_list[int(number[s])]) for s in range(segments_number)]
            
        self.shift = 0
        
    def rotate_right(self, segments_count):
        self.shift += segments_count
        
    def rotate_left(self, segments_count):
        self.shift -= segments_count
        
    def get_segment(self, index):
        return self.segments[(self.shift + index) % self.segments_number]
    
    def set_segment(self, index, value):
        self.segments[(self.shift + index) % self.segments_number] = value
        
    def print_circle(self):
        radius = 10
        thickness = 1
        segments_number = self.segments_number
        steps = 4
        G = Graphics()
        
        for j in range(segments_number):
            L = [[(radius + thickness) * cos(2 * pi * (i + j * steps) / (segments_number * steps)), 
                  (radius + thickness) * sin(2 * pi * (i + j * steps) / (segments_number * steps))] for i in range(steps + 1)]
            L.extend([[radius * cos(2 * pi * (j * steps + steps - i) / (segments_number * steps)), 
                       radius * sin(2 * pi * (j * steps + steps - i) / (segments_number * steps))] for i in range(steps + 1)])
            G += polygon2d(L, axes=False, rgbcolor=self.get_segment(j).color.value, edgecolor="black")
            G += text(str(self.get_segment(j).number), ((radius + thickness / 2) * cos(2 * pi * (1 + j * 2) / (segments_number * 2)), 
                                   (radius + thickness / 2) * sin(2 * pi * (1 + 2 * j) / (segments_number * 2))), rgbcolor="black", fontsize=14)
            G += text(str(j + 1), ((radius - thickness / 2) * cos(2 * pi * (1 + j * 2) / (segments_number * 2)), 
                                   (radius - thickness / 2) * sin(2 * pi * (1 + 2 * j) / (segments_number * 2))), rgbcolor="black", fontsize=14)
            
        G.show(figsize=[16,16])
```

Каждому сектору круга соответствует свой индификационный номер и цвет (номер сектора записан внутри сектора). При повороте круга сектор сохраняет свой номер и цвет. Числа внутри круга идентифицируют места и не меняются при повороте.

```python
number = "0123450123450230444034012301200000002030123040120333303012301223032003"
circle = Circle(len(number), number)

print("Круг для номера {}:".format(number))
circle.print_circle()
```

```python
circle.rotate_right(20)
print("Повернем круг на 20 секторов по часовой стрелке: ")
circle.print_circle()
```
