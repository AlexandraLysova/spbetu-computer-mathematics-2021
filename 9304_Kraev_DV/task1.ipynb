## Основные теоретические положения

### Бинарный метод
Бинарный(двоичный) метод возведения в степень — это метод, позволяющий возводить любое число в степень $n\in N$ за O(log n) умножений (вместо n умножений при обычном подходе). В этом методе мы проходим по степени n в двоичном виде слева направо, начинаем с самого левого бита, и при каждом переходе на следующий бит возводим число в квадрат, если бит, на который мы перешли, равен 1, то после возведения в квадрат, еще и умножаем на исходное число. В итоге, с помощью этого метода мы возвели число в n-ю степень за $\lambda (n)+\nu (n)-1$ умножений($\lambda (n) = \lceil log_2(n) \rceil$, $\nu (n)$  - количество единиц в двоичной записи числа).


### Метод множителей
Метод множителей - метод быстрого возведения в степень, путем разложения степени $n\in N$ на 2 множителя. После разложения происходит возведение числа в степень, равную меньшему множителю. Затем итоговое число возводится в степень, равную второму множителю. Если степень n - простое число, то сначала с помошью этого метода возводим число в степень n-1, а потом домножаем на исходное число. 

###  Аддитивная цепочка
Аддитивной цепочкой для некоторого числа $n\in N$ называют последовательность натуральных чисел до $a_m$, обладающих свойствами: 
$a_i=a_j+a_k,\ k\le j\le i,\ \forall \ i=1\ ...\ m\\
1=a_0,a_1,a_2\ ...\ a_m=n\\ 
a_i=a_j+a_k \ k\le j<i,\ \forall \ i=1\ ...\ m$

Звездной аддитивной цепочкой называется такая аддитивная цепочка, что $a_i = a_{i-1}+a_k$.

### Алгоритм Яо для вычисления приближенных аддитивных цепочек
Цепочка Яо - аддитивная цепочка, состаящая из 2 аддитивных цепочек, первая цепочка равна $Y_k(n):1,2,4,\ ...\, 2^{\lambda (n)}$, вторая аддитивная цепочка составляется на основе первой.\
$d(z)$ - сумму всех разрядов, цифра которых равняется z.\
Алгоритм Яо составляет для числа $n\in N$ цепочку Яо. Сначала алгоритму подается число $n\in N$ и $k\in N$. Далее алгоритм разбивает число на сумму: $\sum_{i=0}^{2^k-1}{2^{ik}a_i}$, где $a_i$ - цифра i-го разряда числа. Потом алгоритм считает $zd(z)$ для всех чисел от 1 до $2^k-1$. $zd(z)$ получается с помощью сложения чисел из $Y_k(n)$. Наконец сложением всех $d(z)$ получаем нужное нам число n. Все промежуточные значения записывются во вторую аддитивную цепочку цепочки Яо.

### Алгоритм дробления вектора индексов
Вектор индексов - последовательность натуральных числе длиной m-1, по которой можно построить аддитивную звездную цепочку длиной m. Каждый элемент в векторе - это индекс элемента звездной цепочки, который нужно прибавить к i-му элементу, чтобы получить i+1 элемент.\
Данный алгоритм нужен для того, чтобы найти для заданного числа $n\in N$ минимальную звездную цепочку. Алгоритм сначала с помощью внешнего цикла перебирает все длины m вектора индексов, где $\underline{l}(n) \le m \le \overline{l}(n)$.  $\underline{l}(n) = \lceil lbn \rceil$, $\overline{l}(n) = \lambda(n) + \nu(n)-1$. Далее происходит деление вектора индексов на 2 части: $\{r_i\}_{i=1}^q$ и $\{\rho\}_{j=q+1}^m$, где $1\le q\le m-1$. Далее с помощью внутреннего цикла происходит перебор первой части вектора, в нем считается максимальное $a_{max}$ и минимальное значение $a_{min}$ итогового числа. Если $a_m \in [a_{min};a_{max}]$ , то с помощью второго внутреннего цикла происходит перебор всех 2 частей вектора, в ином случае алгоритм переходит на следующий вектор. Если $a_m = n$, то алгоритм завершает свою работу. 

### Гипотеза Шольца-Брауэра
Гипотеза Шольца-Брауэра для звездной цепочки заключается в $l^*(n^2-1)\le l^*(n)+n-1$, где $l^*(n)$ - длина минимальной аддитивной цепочки для числа n.

### 3. Алгоритм дробления вектора индексов

```python
def nu(number):
    count = 0
    while number != 0:
        count += number % 2
        number = number // 2
    return count

def getUpperBound(number):
    return (nu(number) + lamb(number) -1) 
    
    

def getLowerBound(number):
    addition = 0
    if number % 1 != 0:
        addition = 1
    return (lamb(number) + addition)


def createChain(vector):
    chain = [1]
    for index in vector:
        chain.append(chain[index-1] + chain[-1])
    return chain

def decVec(vec, q, m):
    maxVec = [a for a in range(q,m+1)]
    minVec = [1 for a in range(q,m+1)]
    zeroVec = [0 for a in range(q,m+1)]
    if vec == minVec:
        return zeroVec
    ind = 1
    while (vec[-ind] == 1) and (ind < len(vec)):
        vec[-ind] = maxVec[-ind]
        ind += 1
    vec[-ind] -= 1
    return vec
            

def splitting(number):
    if number == 1:
        return [1]
    if number == 2:
        return [1,2]
    lowerBound = getLowerBound(number)
    upperBound = getUpperBound(number)
    for m in range(lowerBound, upperBound+1):
        chain = []
        q = m//2
        vec1 = [a for a in range(1,q+1)]
        stopvec1 = [0 for a in range(q)]
        while vec1 != stopvec1:
            chain = createChain(vec1)
            amax = chain[-1] * 2^(m-q)
            amin = chain[-1] + m - q
            stopvec2 = [0 for a in range(m-q)]
            vec2 = [a for a in range(q+1,m+1)]
            finChain = []
            while vec2 != stopvec2:
                if(amin > number or amax < number):
                    break
                finChain = createChain(vec1 + vec2)
                
                if finChain[-1] == number:
                    return finChain
                vec2 = decVec(vec2,q+1,m)
            vec1 = decVec(vec1,1,q)

```

```python
import time
headers = ['Число','Аддитивная цепочка','Время поиска(сек)']
L = []
line = []

t0 = time.time()
chain = splitting(599)
t1 = time.time()
line = [str(599),str(chain), str(t1 - t0)]
L.append(line)

fig, ax = plt.subplots()
    
ax.set_axis_off() 
table = ax.table( 
        cellText = L,   
        colLabels = headers,   
        cellLoc ='center',  
        loc ='upper left'
    
        )
table.auto_set_font_size(False)
cellDict = table.get_celld()
for i in range(2):
    cellDict[(i,0)].set_width(0.2)
    cellDict[(i,1)].set_width(0.7)
    cellDict[(i,2)].set_width(0.4)
        
ax.set_title('Алгоритм дробления индексов для числа 599',fontsize=15,fontweight='bold',loc='left') 
table.set_fontsize(14)
table.scale(2,3)

plt.show()
```

Алгоритм дробления вектора индексов находит минимальную аддитивную звуздную цепочку для заданного числа, в отличие от алгоритма Яо, который находит только приближенную, однако, алгоритм Яо работает намного быстрее.

### 4. Проверка следствия 1 теоремы Брауэра

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import leastsq

def residuals(coeffs, y, t):
    return  y - np.log(t)*coeffs[0]

x = []
y = []
for i in range(1,201):
    x.append(i)
    y.append(len(splitting(i)))

figure, axes = plt.subplots(figsize=(10,10))
p, flag = leastsq(residuals, 1, args=(y, x))
plt.plot(x,y,'o',x, np.log(x)*p[0])
plt.show()
```

Из графика видно, что функция l(n) аппроксимируется логарифмической прямой. То есть при увеличении n значение l(n) будет изменяться незначительно, как логарифмическая функция. Это подтверждает 1 следствие теоремы Брауэра, что $\lim_{n\to\infty}\frac{l(n)}{\lambda(n)}=1$.

### 5. Проверка гипотезы Шольца-Брауэра

```python
x = list(range(1,21))
f_star1 = []
f_star2 = []
for i in range(1,21):
    f_star1.append(len(splitting(i**2)))
    f_star2.append(len(splitting(i)) + i -1)
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,f_star1,'red',label="l(n^2-1)")
ax.plot(x,f_star2,'blue',label="l(n)+n-1")
ax.legend()
ax.grid()
plt.show()
    
```

Из графика видно, что $l(n^2-1)$ всегда больше либо равно $l(n)+n-1$, следовательно гипотеза Шольца-Брауэра верна.
